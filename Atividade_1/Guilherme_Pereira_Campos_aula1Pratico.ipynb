### Exercício 1: Carregar e salvar o dataset
Abra o arquivo "Iris" do UCI Repository. Carregue o dataset no seu código usando a biblioteca
pandas e salve-o em um DataFrame: https://archive.ics.uci.edu/ml/datasets/iris

#### Resolução 1: 
- 1. Importar a biblioteca pandas.  
- 2. Passar a `URL` do dataset. 
- 3. Ler o dataset usando a função `pd.read_csv` do Pandas passando a url e os nome das colunas. 


```python
# Importando  Pandas 
import pandas as pd 

# URL do Dataset 
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Salvando em um dataframe Pandas 
df_iris = pd.read_csv(url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
# Importando o display para melhor visualização
from IPython.display import display 

# Exibindo o número de linhas e colunas 
print("Número de linhas e colunas:")
print(f"linhas: {df_iris.shape[0]}, colunas: {df_iris.shape[1]}")

print('--------------------------------------------------')

# Exibindo o tipo de dado de cada coluna usando .info()
print("Tipo de dado de cada coluna:")
print(df_iris.info())

print('--------------------------------------------------')

# Exibindo estatísticas descritivas usando .describe()
print("Estatísticas descritivas:")
display(df_iris.describe())
```

### Exercício 3: Gráficos
Gere os seguintes gráficos usando a biblioteca matplotlib ou seaborn:
- Um histograma para cada uma das colunas numéricas.
- Um boxplot para as colunas numéricas, segmentadas por espécie.
- Um scatter plot (diagrama de dispersão) relacionando pares de colunas.

#### Resolução 3: 

- 1. Importar o Matplotlib e o Seaborn.
- 2. Configurar alguns parâmetros do Seaborn com o `sns.set`. 
- 3. Criar uma figura com o Matplotlib com o tamanho 25x30. 
- 4. Subplots do Matplotlib: Os subplots  tem como formato (`linhas`, `colunas`, `gráfico_atual`), dessa forma é possível fazer uma figura com vários gráficos. 
- 5. Os histogramas são plotados com os subplots de 1 a 4, usando o `sns.histplot` com o parâmetro `kde=True` para visualizar a estimativa de densidade. 
- 6. Os boxplots são plotados com os subplots de 5 a 8, usando a segmentação por espécie por meio do parâmetro `hue`.
- 7. Os scatterplots são plotados com os subplots de 9 a 14, relacionando os pares de colunas. 
- Todos gŕaficos foram configurados com título, rótulo nos eixos, legenda, cor, tamanho e peso de fonte. 

```python
# Importando o seaborn e o matplotlib 
import seaborn as sns
import matplotlib.pyplot as plt 

# Configuração do Seaborn 
sns.set(
    style='darkgrid',
    context='notebook',
    rc={
        'axes.spines.top': False,
        'axes.spines.right': False
    }
) 

# Cria uma figura com o plt.figure de tamanho 25x30
plt.figure(figsize=(25,30))

# Primeiro histograma
plt.subplot(4, 4, 1)
sns.histplot(x = df_iris['sepal_length'], kde = True, color='#8aad78')
plt.title('Primeiro Histograma', fontsize=14, fontweight='bold')
plt.ylabel('Contagem', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento da sépala', fontsize=12, fontweight='bold')

# Segundo histograma
plt.subplot(4, 4, 2)
sns.histplot(x=df_iris['sepal_width'], kde = True, color='#789aad')
plt.title('Segundo Histograma', fontsize=14, fontweight='bold')
plt.ylabel('Contagem',fontsize=12, fontweight='bold')
plt.xlabel('Largura da sépala', fontsize=12, fontweight='bold')

# Terceiro histograma
plt.subplot(4, 4, 3)
sns.histplot(x=df_iris['petal_length'], kde = True, color='#ad7879')
plt.title('Terceiro Histograma', fontsize=14, fontweight='bold')
plt.ylabel('Contagem', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento da pétala', fontsize=12, fontweight='bold')

# Quarto histograma
plt.subplot(4, 4, 4)
sns.histplot(x=df_iris['petal_width'], kde = True, color='#787dad')
plt.title('Quarto Histograma', fontsize=14, fontweight='bold')
plt.ylabel('Contagem', fontsize=12, fontweight='bold')
plt.xlabel('Largura da pétala', fontsize=12, fontweight='bold')

# Primeiro boxplot  
plt.subplot(4, 4, 5) 
sns.boxplot(data=df_iris, x='class', y='sepal_length', hue='class', linecolor='black', palette=['#b5675b','#5bb573', '#5b67b5'])
plt.title('Primeiro Boxplot', fontsize=14, fontweight='bold')
plt.ylabel('Comprimento de sépala', fontsize=12, fontweight='bold')
plt.xlabel('Espécie', fontsize=12, fontweight='bold')

# Segundo boxplot 
plt.subplot(4, 4, 6) 
sns.boxplot(data=df_iris, x='class', y='sepal_width', hue='class', linecolor='black', palette=['#898f77', '#778f81', '#8f7777'])
plt.title('Segundo Boxplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura de sépala', fontsize=12, fontweight='bold')
plt.xlabel('Espécie', fontsize=12, fontweight='bold')

# Terceiro boxplot 
plt.subplot(4, 4, 7) 
sns.boxplot(data=df_iris, x='class', y='petal_length', hue='class', linecolor='black', palette=['#bfc25f', '#6b5fc2', '#c25f5f'])
plt.title('Terceiro Boxplot', fontsize=14, fontweight='bold')
plt.ylabel('Comprimento da pétala', fontsize=12, fontweight='bold')
plt.xlabel('Espécie', fontsize=12, fontweight='bold')

# Quarto boxplot 
plt.subplot(4, 4, 8) 
sns.boxplot(data=df_iris, x='class', y='petal_width', hue='class', linecolor='black', palette=['#617351', '#515d73', '#6c5173'])
plt.title('Quarto Boxplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura da pétala', fontsize=12, fontweight='bold')
plt.xlabel('Espécie', fontsize=12, fontweight='bold')

# Primeiro scatterplot 
plt.subplot(4, 4, 9)
sns.scatterplot(data=df_iris, x='sepal_length', y='sepal_width', hue='class', palette='viridis', s=50)
plt.title('Primeiro Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura de sépala', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento de sépala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best')

# Segundo scatterplot 
plt.subplot(4, 4, 10)
sns.scatterplot(data=df_iris, x='sepal_length', y='petal_length', hue='class', palette='deep', s=50)
plt.title('Segundo Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura da sépala', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento da pétala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best')

# Terceiro scatterplot 
plt.subplot(4, 4, 11)
sns.scatterplot(data=df_iris, x='sepal_length', y='petal_width', hue='class', palette=['#a88732', '#32a842', '#3263a8'], s=50)
plt.title('Terceiro Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura da pétala', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento da sépala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best') 

# Quarto scatterplot 
plt.subplot(4, 4, 12)
sns.scatterplot(data=df_iris, x='sepal_width', y='petal_width', hue='class', palette=['#3b5c03', '#035c52', '#5c0303'], s=50)
plt.title('Quarto Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('Largura de pétala', fontsize=12, fontweight='bold')
plt.xlabel('Largura de sépala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best') 

# Quinto scatterplot 
plt.subplot(4, 4, 13)
sns.scatterplot(data=df_iris, x='sepal_width', y='petal_length', hue='class', palette=['#b3a78f', '#8fb396', '#b38f8f'], s=50)
plt.title('Quinto Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('Comprimento de pétala', fontsize=12, fontweight='bold')
plt.xlabel('Largura de sépala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best') 

# Sexto scatterplot 
plt.subplot(4, 4, 14)
sns.scatterplot(data=df_iris, x='petal_length', y='petal_width', hue='class', palette=['#f7c80a', '#0a49f7', '#f70a0a'], s=50)
plt.title('Sexto Scatterplot', fontsize=14, fontweight='bold')
plt.ylabel('largura de pétala', fontsize=12, fontweight='bold')
plt.xlabel('Comprimento de pétala', fontsize=12, fontweight='bold')
plt.legend(title='Espécie', fontsize='medium', loc='best') 
```
